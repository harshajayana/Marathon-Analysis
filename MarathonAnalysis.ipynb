```python
df2=df[(df['Event distance/length'].isin(['50km', '50mi'])) &
    (df['Year of event'] == 2020) &
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
#remove (USA) from the event name
```

```python
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up Athelete age
```

```python
df2['Athlete year of birth'].fillna(0, inplace=True)

# Ensure 'Athlete year of birth' is of integer type
df2['Athlete year of birth'] = df2['Athlete year of birth'].astype(int)

# Calculate the age in the year 2020
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']

# Display the DataFrame to verify the new column
print(df2[['Athlete year of birth', 'Athlete_age']])
```

```python
#remove h from athelete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.replace('h', '').str.strip()
```

```python
df2.head(5)
```

```python
df2= df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```
